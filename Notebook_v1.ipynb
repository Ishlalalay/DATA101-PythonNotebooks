```python
# If your date column is not yet in datetime format, convert it
df['datetime'] = pd.to_datetime(df['datetime'])

# Extract just the date into a new column
df['date'] = df['datetime'].dt.date

# Filter rows for a specific date
target_date = pd.to_datetime("2025-12-01").date()
filtered_df = df[df['date'] == target_date]
```

```python
'''
import json
import topojson

with open("philippines-json-maps-master./2023/geojson/country/medres/country.0.01.json") as f:
    topo = json.load(f)

# Convert to GeoJSON
geojson = topojson.Topology(topo).to_geojson()

with open("philippines_country.geojson", "w") as f:
    f.write(geojson)
'''

import geopandas as gpd
import os

folder = "philippines-json-maps-master./2023/geojson/municities/lowres"
geo_dfs = []

for file in os.listdir(folder):
    if file.endswith(".json") or file.endswith(".geojson"):
        gdf = gpd.read_file(os.path.join(folder, file))
        geo_dfs.append(gdf)

# Combine into one GeoDataFrame
ph_cities = gpd.GeoDataFrame(pd.concat(geo_dfs, ignore_index=True))
```

```python
print(ph_cities.columns)   # columns from your GeoJSON
print(df.columns)          # columns from your AQI dataframe
```

```python
import geopandas as gpd
import folium

merged = ph_cities.merge(filtered_df, left_on="adm4_en", right_on="city_name")

merged = merged[['city_name', 'main.aqi', 'geometry']]

m = folium.Map(location=[12.8797, 121.7740], zoom_start=6)

folium.Choropleth(
    geo_data=merged,
    data=merged,
    columns=["city_name", "main.aqi"],   # match the actual names
    key_on="feature.properties.city_name",
    fill_color="YlOrRd",
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name="Air Quality Index (AQI)"
).add_to(m)

# Add tooltips for city names and AQI
folium.GeoJson(
    merged,
    style_function=lambda x: {"fillColor": "transparent", "color": "black", "weight": 0.5},
    tooltip=folium.GeoJsonTooltip(fields=["city_name", "main.aqi"])
).add_to(m)

# Save map to HTML
m.save("philippines_aqi_map.html")
```
